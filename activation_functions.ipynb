**Activation functions**

**Sigmoid**

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-10,10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('activation function:sigmoid')
plt.show()
```

**Activation function tanh**

```python
 def tanh(x):
  return np.tanh(x)
  return 1/(1+np.exp(-x))
x=np.linspace(-19,19)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('activation function:tanh')
plt.show()
```

**Activation function reLU**

```python
import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
        x1.append(i)
  return x1

x=np.linspace(-10,10)
plt.plot(x, ReLU(x))
plt.axis('tight')
plt.title('activation function:ReLU')
plt.show()
```
